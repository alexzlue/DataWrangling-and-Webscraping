```python
# I got my data from: https://data.world/cdc/children-with-current-asthma
# Topic: Asthma in the US
# read_csv fcn arguments used: header, skipfooter, engine, usecols, skiprows, names, nrows

#open the csv with pandas
import pandas as pd
dataset = "./Asthma_dataset.csv"
df = pd.read_csv(dataset)

#now display raw dataframe
df
```

```python
# The data we have is split by percentage, so lets make row 0 our header instead of the current
mod_df= pd.read_csv(dataset, header=1)
# ^ this makes our header the first row of our csv, which is "characteristic" and the years.

mod_df
```

```python
# there is a lot of unnecessary rows (29-33) at the bottom of our data set. rather than 
# manually manipulating the dataset, lets remove them. We can remove by using skipfooter.
# This will skip the last n rows. Since we want to skip 29-33, we can use
removed_footer_df = pd.read_csv(dataset, header=1, skipfooter=5, engine='python')
# since we are skipping the last 5 rows of the csv. We will need to specify the engine to be python, however,
# since it is unsupported in the 'c' parser

removed_footer_df
```

```python
# lets only take a look at the data from the 2002 and 2012 to analyze the difference in one decade
# we will use 'usecols' argument and read columns 2 and 12 for the years and row 0 for the Characteristic
decade_df = pd.read_csv(dataset, header=1, skipfooter=5, engine='python', usecols=[0, 2, 12])

decade_df
```

```python
# we have data rows that separate the demography (ie row 5 in df that shows race/ethnicity for everything underneath it)
# lets use skiprows to isolate those specific rows and lets change the header from 'characteristic' to 'race/ethnicity'

naming = [i for i in range(2001, 2014)]
naming.insert(0,"Race/Ethnicity")
# these will allow us to use our own naming scheme for the header. We have "Race/Ethnicity" at the beginning followed
# by 2001 to 2013 for the years

skip_after = [i for i in range(14,35)] #this captures all the row# for rows that appear after race/ethnicity entries
skip_before = [i for i in range(0,7)] #this captures all the row# for rows that appear before race/ethnicity entries
race_eth_df = pd.read_csv(dataset, names=naming, skiprows=skip_after+skip_before)

race_eth_df
```

```python
# so we know there are 7 race/ethnicities to consider, we can use that to output a data frame that only 
# reads the first 7 rows of each category with 'nrows'. Let's use this to display the female demographics

naming = [i for i in range(2001, 2014)]
naming.insert(0,"Females: Race/Ethnicity")
# these will allow us to use our own naming scheme for the header. We have "females: Race/Ethnicity" at the beginning followed
# by 2001 to 2013 for the years

skip_before = [i for i in range(0,23)] # skip all rows before the female category.

f_race_eth_df = pd.read_csv(dataset, names=naming, skiprows=skip_before, nrows=7)
#using nrows=7 will read the only 7 rows, after the skiprows take place

f_race_eth_df
```
